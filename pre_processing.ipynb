```python
from PIL import Image
import os
import numpy as np
from collections import Counter

def calculate_color_distance(color1, color2):
    r1, g1, b1 = color1
    r2, g2, b2 = color2
    distance = np.sqrt((r1 - r2) ** 2 + (g1 - g2) ** 2 + (b1 - b2) ** 2)
    return distance

def get_most_similar_color(target_colors, pixel_color):
    min_distance = float("inf")
    most_similar_color = None
    
    for target_color in target_colors:
        distance = calculate_color_distance(target_color, pixel_color)
        if distance < min_distance:
            min_distance = distance
            most_similar_color = target_color
    
    return most_similar_color

def replace_colors(image_path, output_path, target_colors):
    image = Image.open(image_path)
    width, height = image.size
    new_image = Image.new("RGB", (width, height))
    
    for y in range(height):
        for x in range(width):
            pixel = image.getpixel((x, y))
            
            if pixel not in target_colors:
                replacement_color = get_most_similar_color(target_colors, pixel)
                new_image.putpixel((x, y), replacement_color)
                # print(replacement_color)
            else:
                new_image.putpixel((x, y), pixel)
                # print(pixel)
    
    new_image.save(output_path)

file_list = sorted(os.listdir('/home/cumtailab227/zrh/zrh/coalseg_chen_0820/gt'))
for file in file_list:
    print(file)
    input_image_path = os.path.join('/home/cumtailab227/zrh/zrh/coalseg_chen_0820/gt', file)
    output_image_path = os.path.join('/home/cumtailab227/zrh/zrh/coalseg_chen_0820/label_v', file.split('.')[0]+'.png')

    target_colors = [(64, 140, 216), (0, 255, 0), (0, 0, 255), (255, 0, 0)]

    replace_colors(input_image_path, output_image_path, target_colors)
```

```python
from PIL import Image
import os
import random

def crop_images(input_folder, output_folder, output_size, num_crops_per_image, seed):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image_files = sorted([f for f in os.listdir(input_folder) if f.endswith((".jpg", ".png", ".jpeg"))])
    print(image_files)

    # 设置相同的随机种子
    random.seed(seed)

    for image_file in image_files:
        image_path = os.path.join(input_folder, image_file)
        image = Image.open(image_path)
        image = image.convert("RGB")  # 将图片格式统一为RGB模式
        image_width, image_height = image.size

        left_quarter_width = 512
        left_quarter_height = image_height
        right_three_quarters_height = image_height

        for _ in range(num_crops_per_image):
            region_choice = random.choices(["val", "train"], weights=[1, 9])[0]

            if region_choice == "val":
                crop_start_x = random.randint(0, left_quarter_width - output_size[0])
                crop_start_y = random.randint(0, left_quarter_height - output_size[1])
            else:
                crop_start_x = random.randint(left_quarter_width, image_width - output_size[0])
                crop_start_y = random.randint(0, right_three_quarters_height - output_size[1])

            print(crop_start_x, crop_start_y)

            cropped_image = image.crop((crop_start_x, crop_start_y, crop_start_x + output_size[0], crop_start_y + output_size[1]))

            output_filename = f"{os.path.splitext(image_file)[0]}_{region_choice}_{crop_start_x}_{crop_start_y}.png"
            output_path = os.path.join(output_folder, output_filename)
            cropped_image.save(output_path)

        image.close()

    print(f"随机裁剪完成: {input_folder} -> {output_folder}")

# 对不同文件夹进行裁剪，确保使用相同的随机种子
num_crops_per_image = 200
output_size = (512, 512)
seed = 2  # 使用相同的随机种子

crop_images('label_v',  'cut_label_v',  output_size, num_crops_per_image, seed)
crop_images('image1',   'cut_image1',   output_size, num_crops_per_image, seed)
crop_images('image2',   'cut_image2',   output_size, num_crops_per_image, seed)
```

```python
import os
from PIL import Image
import cv2
import numpy as np

def map_colors_to_labels(image):
        color_mapping = {(64, 140, 216) : 0,   
                         (0, 255, 0)    : 1,   
                         (0, 0, 255)    : 2, 
                         (255, 0, 0)    : 3}
        
        label_image = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)

        for y in range(image.shape[0]):
            for x in range(image.shape[1]):
                pixel_color = tuple(image[y, x])
                if pixel_color in color_mapping:
                    label_image[y, x] = color_mapping[pixel_color]
                else:
                    print('标签中存在别的颜色！')
        return label_image
            
# 设置输入和输出文件夹路径
input_folder = 'cut_label_v'  # 输入文件夹路径，存放原始图片
output_folder = 'cut_label'  # 输出文件夹路径，用于保存处理后的图片

# 确保输出文件夹存在
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 支持的图片文件扩展名
supported_extensions = ['.jpg', '.jpeg', '.png', '.gif']

# 遍历输入文件夹中的所有文件
for filename in os.listdir(input_folder):
    if any(filename.lower().endswith(ext) for ext in supported_extensions):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        # 打开图片并进行操作，这里只是一个示例，您可以替换为您实际的操作
            # 进行某种操作，比如调整大小、变换颜色等
            # 这里仅示范保存图片，您需要根据您的操作来修改这部分
        label = cv2.cvtColor(cv2.imread(input_path), cv2.COLOR_BGR2RGB)
        label = map_colors_to_labels(label)

        # 保存处理后的图片到输出文件夹
        cv2.imwrite(output_path, label)

print("操作完成！")
```
